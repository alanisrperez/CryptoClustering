```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(5)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
market_data_scaled = StandardScaler().fit_transform(df_market_data)
```

```python
# Create a DataFrame with the scaled data
df_market_data_scaled = pd.DataFrame(market_data_scaled, columns=['price_change_percentage_24h','price_change_percentage_7d','price_change_percentage_14d','price_change_percentage_30d','price_change_percentage_60d','price_change_percentage_200d','price_change_percentage_1y'])

# Copy the crypto names from the original DataFrame
df_market_data_scaled['coin_id'] = df_market_data.index

# Set the coin_id column as index
df_market_data_scaled = df_market_data_scaled.set_index('coin_id')

# Display the scaled DataFrame
df_market_data_scaled
```

---

### Find the Best Value for k Using the Original Scaled DataFrame.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
for i in k:
    
# 1. Create a KMeans model using the loop counter for the n_clusters
    k_model = KMeans(n_clusters=i, random_state=1)
    
# 2. Fit the model to the data using `df_market_data_scaled`
    k_model.fit(df_market_data_scaled)
    
# 3. Append the model.inertia_ to the inertia list
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {'k': k, 'inertia': inertia}

# Create a DataFrame with the data to plot the Elbow curve
elbow_df = pd.DataFrame(elbow_data)
elbow_df
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
original_scaled_elbow_df = elbow_df.hvplot.line(
    x =  'k',
    y = 'inertia',
    label = 'Original Scaled Data'
)

original_scaled_elbow_df
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** Either 3 or 4 could be good values for k in this case. I am choosing to select 4 for k because the curve has an additional small divot in the line before evening out, and I would like to see if a value of 4 can show us something additional in our data that a value of 3 may not show.

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialize the K-Means model using the best value for k
model = KMeans (n_clusters = 4, random_state=1)
```

```python
# Fit the K-Means model using the scaled DataFrame
model.fit(df_market_data_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
k_4 = model.predict(df_market_data_scaled)

# Print the resulting array of cluster values.
k_4
```

```python
# Create a copy of the scaled DataFrame
df_market_data_predictions = df_market_data_scaled.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
df_market_data_predictions['spread_4'] = k_4

# Display the copy of the scaled DataFrame
df_market_data_predictions
```

```python
# Create a scatter plot using hvPlot by setting
original_scaled_plot = df_market_data_predictions.hvplot.scatter(
    
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
    x = 'price_change_percentage_24h',
    y = 'price_change_percentage_7d',
    
# Color the graph points with the labels found using K-Means and
    by = 'spread_4',
    
# add the crypto name in the `hover_cols` parameter to identify
 # the cryptocurrency represented by each data point.   
    hover_cols = ['coin_id'],
    label = 'Original Scaled Data'
)

original_scaled_plot
```

---

### Optimize Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled DataFrame
# down to three principal components.
market_data_pca = pca.fit_transform(df_market_data_scaled)

# View the scaled PCA data
market_data_pca[:5]
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
print(f"Explained Variance Ratio: {pca.explained_variance_ratio_}")
print(f"Sum of Explained Variance Ratio: {sum(pca.explained_variance_ratio_)}")
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** The first principal component accounts for 37% of the data, the second component accounts for 35%, and the third accounts for 18%. The first two components account for a significant portion of the variability, whereas the third component contributes less influence. The combined sum of the explained variance ratio for all components is approximately 89.5%, which tells us that together they encompass a considerate amount of the variability in the dataset. This high total explained variance indicates that these three components are effective in representing the data.

```python
# Create a new DataFrame with the PCA data.
df_market_data_pca = pd.DataFrame(market_data_pca, columns=['PC1', 'PC2', 'PC3'])

# Copy the crypto names from the original scaled DataFrame
df_market_data_pca['coin_id'] = df_market_data_scaled.index

# Set the coin_id column as index
df_market_data_pca = df_market_data_pca.set_index('coin_id')

# Display the scaled PCA DataFrame
df_market_data_pca
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1,11))
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
for i in k:
# 1. Create a KMeans model using the loop counter for the n_clusters
    k_model_pca = KMeans(n_clusters = i, random_state = 1)
# 2. Fit the model to the data using `df_market_data_pca`
    k_model_pca.fit(df_market_data_pca)
# 3. Append the model.inertia_ to the inertia list
    inertia.append(k_model_pca.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_pca_data = {'k': k, 'inertia': inertia}

# Create a DataFrame with the data to plot the Elbow curve
elbow_pca_df = pd.DataFrame(elbow_pca_data)
elbow_pca_df
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
pca_scaled_elbow_df = elbow_pca_df.hvplot.line(
    x = 'k',
    y = 'inertia',
    label = 'PCA Scaled Data'
)

pca_scaled_elbow_df
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** 4


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** No. Although 3 can be a good representative value of k, such as we saw with the original scaled data, I am choosing to go forward with 4 as the line follows an even and straight path with little curve as it continues.

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialize the K-Means model using the best value for k
model_pca = KMeans(n_clusters = 4, random_state = 1)
```

```python
# Fit the K-Means model using the PCA data
model_pca.fit(df_market_data_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
pca_k_4 = model_pca.predict(df_market_data_pca)

# Print the resulting array of cluster values.
pca_k_4
```

```python
# Create a copy of the scaled PCA DataFrame
df_market_data_pca_predictions = df_market_data_pca.copy()

# Add a new column to the copy of the PCA DataFrame with the predicted clusters
df_market_data_pca_predictions['spread_4'] = pca_k_4

# Display the copy of the scaled PCA DataFrame
df_market_data_pca_predictions
```

```python
# Create a scatter plot using hvPlot by setting
pca_scaled_plot = df_market_data_pca_predictions.hvplot.scatter(
# `x="PC1"` and `y="PC2"`.
    x = 'PC1',
    y = 'PC2',
# Color the graph points with the labels found using K-Means and
    by = 'spread_4',
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.   
    hover_cols = 'coin_id',
    label = 'PCA Scaled Data'
)

pca_scaled_plot
```

  * **Answer:** To ensure that clusters are separated in a way that is meaningful and accurately representative of the data while at the same time being manageable for analysis/interpretation, the choice in selecting a proper k value can be intimidating. Using fewer features can contribute to underfitting the data, to the point where the model is not detailed enough to show the structure of the data set. If clusters are too broad or lack detail, then they may not represent the data well. Lower k values can give higher variance, implying that data points are too similar to interpret well. Although interpretation may be easier, this can oversimply our model, with doesn't give us a lot of insight, or may generalize the data, leaving out the opportunity for complexities and potential findings. In our case here, a k value of 4 gave us insight into some data points that may qualify as outliers or may carry some complexity that the other data points do not follow.
